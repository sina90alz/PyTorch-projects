## PyTorch Tensor and NumPy

NumPy is a scientific numerical computing library.

And PyTorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.Numpy()`

```python
 # Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Attention to the type difference, numpy is by default 64, torch is 32
array.dtype, tensor.dtype, torch.arange(1.0, 8.0).dtype
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

## Reproductubility (Trying to take random out of random)

How neural network learns

 `start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again ...`

 To reduce randomness in neural networks and PyTorch comes the concept of **random seed**.
 Essentially what the random seed does is "flavour the randomness ."

```python
torch.rand(3,3)
```

```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)

print(random_tensor_A == random_tensor_B)
```

```python
# Let's make random but reproductible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)

print(random_tensor_C == random_tensor_D)
```
